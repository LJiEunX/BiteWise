```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("caps_fin.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df = df.drop(columns=['Unnamed: 13', 'Unnamed: 14'])
df.info()
df.head()
```

```python
df.describe()
```

```python
jakarta_utara_data = df[df['daerah'] == 'Jakarta Utara']

# Remove rows with NaN in 'rata_rata' and convert 'rata_rata' to numeric
jakarta_utara_data = jakarta_utara_data.dropna(subset=['rata_rata'])
jakarta_utara_data['rata_rata'] = jakarta_utara_data['rata_rata'].replace('[Rp,]', '', regex=True).astype(float)

# Calculate the average of 'rata_rata' for Jakarta Barat
average_price_jakarta_utara = jakarta_utara_data['rata_rata'].mean()
average_price_jakarta_utara
```

#rata rata rating pada setiap daerah

```python
total_rate = df.groupby([ 'daerah'])['rating'].mean()

# Menampilkan total penyewa sepeda pada musim 1 per hari dengan menampilkan hari dalam seminggu
print("Rata rata rating:")
print(total_rate)
plt.figure(figsize=(10, 6))
total_rate.plot(kind='bar', color='skyblue')
plt.title('Rata rata rating perdaerah')
plt.xlabel('Daerah')
plt.ylabel('RatingTotal')
plt.xticks(rotation=0)  # Rotasi label x agar lebih mudah dibaca
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
tipe_perbandingan = df.groupby(['daerah', 'tipe']).size().unstack(fill_value=0)

# Displaying the data in tabular form
print(tipe_perbandingan)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
tipe_perbandingan.plot(kind='bar', stacked=True, colormap='viridis', ax=plt.gca())
plt.title('Total Banyaknya Tipe pada Setiap Daerah')
plt.xlabel('Daerah')
plt.ylabel('Jumlah Tipe')
plt.xticks(rotation=0)  # Rotate x labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Tipe')
plt.tight_layout()
plt.show()
```

```python
# Calculate average prices for each region
average_price_jakarta_pusat = df[df['daerah'] == 'Jakarta Pusat']['rata_rata'].replace('[Rp,]', '', regex=True).astype(float).mean()
average_price_jakarta_timur = df[df['daerah'] == 'Jakarta Timur']['rata_rata'].replace('[Rp,]', '', regex=True).astype(float).mean()
average_price_jakarta_barat = df[df['daerah'] == 'Jakarta Barat']['rata_rata'].replace('[Rp,]', '', regex=True).astype(float).mean()
average_price_jakarta_selatan = df[df['daerah'] == 'Jakarta Selatan']['rata_rata'].replace('[Rp,]', '', regex=True).astype(float).mean()
average_price_kepulauan_seribu = df[df['daerah'] == 'Kepulauan Seribu']['rata_rata'].replace('[Rp,]', '', regex=True).astype(float).mean()

# Prepare data for plotting
regions = ['Jakarta Pusat', 'Jakarta Timur', 'Jakarta Barat', 'Jakarta Selatan', 'Kepulauan Seribu']
average_prices = [
    average_price_jakarta_pusat, 
    average_price_jakarta_timur, 
    average_price_jakarta_barat, 
    average_price_jakarta_selatan, 
    average_price_kepulauan_seribu
]

# Plot the data
plt.figure(figsize=(10, 6))
plt.bar(regions, average_prices, color=['blue', 'green', 'red', 'purple', 'orange'])
plt.xlabel('Daerah')
plt.ylabel('Rata-rata Harga (Rp)')
plt.title('Rata-rata Harga Makanan di Berbagai Daerah di Jakarta')
plt.xticks(rotation=45)
plt.show()
```

```python
correlation_matrix = df[['harga_akhir', 'rating']].corr()

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 12})
plt.title('Correlation between Food Price and Restaurant Rating')
plt.show()
```

```python
areas = ['Jakarta Utara', 'Jakarta Selatan', 'Jakarta Timur', 'Jakarta Barat', 'Jakarta Pusat', 'Kepulauan Seribu']
type_counts_list = []

for area in areas:
    area_data = df[df['daerah'] == area]
    type_counts = area_data['tipe'].value_counts()
    type_counts_list.append(type_counts)

# Create subplots for each area
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

for i, ax in enumerate(axes.flat):
    if i < len(areas):
        ax.pie(type_counts_list[i], labels=type_counts_list[i].index, autopct='%1.1f%%', startangle=140)
        ax.set_title(f'Distribution of Restaurant Types in {areas[i]}')
        ax.axis('equal')

plt.tight_layout()
plt.show()
```
